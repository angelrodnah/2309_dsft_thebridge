4. Llama al eje X "eje X", al eje Y "eje Y" y pon de títulos "recta sube" y "recta baja". Muestra dos etiquetas de ejeX pero solo una de eje Y.

```python
plt.figure()

plt.subplot(1,2,1)
y1 = 3*x
plt.plot(x, y1)
plt.xlim(0,5)
plt.ylim(-15,15)
plt.xlabel("Eje x")
plt.ylabel("Eje y")
plt.title("Recta sube")

plt.subplot(1,2,2)
y2 = -3*x - 5
plt.plot(x, y2)
plt.xlim(0,5)
plt.ylim(-15,15)
plt.xlabel("Eje x")
# plt.ylabel("Eje y")
plt.title("Recta baja")
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es más sencillo dejarlos dentro de la figura.

```python
personas = ["Ana","Luis","Pedro","Sofía","Carmen"]
alturas = [160, 180, 175, 190, 170]

plt.bar(personas, alturas, color="green")
plt.xlabel("Personas")
plt.ylabel("Alturas (cm)")
plt.title("Comparación de alturas")
```

```python
personas = ["Ana","Luis","Pedro","Sofía","Carmen"]
alturas = [160, 180, 175, 190, 170]
df = pd.DataFrame({'alturas':alturas}, index=personas).sort_values(['alturas'], ascending=False)

plt.bar(df.index, df.alturas, color="green")
plt.xlabel("Personas")
plt.ylabel("Alturas (cm)")
# plt.yticks([])
plt.title("Comparación de alturas")
plt.ylim(150,200)
```

13. Vamos a introducir pandas. Crea un DataFrame con las columnas "Year" de valores 2015, 2016, 2017, 2018, 2019 y la columna "Sold_items_A" de valores 1000, 3500, 4000, 5500, 7000

```python
df_13 = pd.DataFrame({"Year": [2015,2016,2017,2018,2019],
                      "Sold_items_a": [1000, 3500, 4000, 5500, 7000]})
df_13
```

14. Ahora pintamos un gráfico de línea con las ventas respecto al año. Pon un título y etiquetas en los ejes.
Cambia el estilo a 'seaborn-white'. La línea debe ser con rayas y verde.

```python
plt.style.use("seaborn-v0_8-white")

plt.figure(figsize=(8,5))
plt.plot(df_13['Year'], df_13['Sold_items_a'], "--g")
plt.xticks(df_13['Year']);
plt.xlabel("Años")
plt.ylabel("Ventas (A)")
plt.title("Evolución de las ventas del departamento A");
```

15. Otro departamento B ha vendido en esos años 2000, 3100, 5000, 4000, 6000 unidades. Incluye esa columna en el DataFrame y pinta en la misma gráfica las dos líneas. B es una línea punteada y roja. Muestra una leyenda abajo a la derecha.

```python
df_13['Sold_items_b'] = [2000, 3100, 5000, 4000, 6000]
df_13
```

```python
plt.figure(figsize=(8,5))
plt.plot(df_13['Year'], df_13['Sold_items_a'], "--g", label="Departamento A")
plt.plot(df_13['Year'], df_13['Sold_items_b'], ":r", label="Departamento B")
plt.xticks(df_13['Year']);
plt.xlabel("Años")
plt.ylabel("Ventas (€)")
plt.title("Evolución de las ventas por departamento")
plt.legend(loc="lower right", fontsize=10);
```
